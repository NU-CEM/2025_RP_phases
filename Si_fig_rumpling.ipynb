```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplpub
mplpub.setup(template='acs')
```

```python
n1_z_disp = pd.read_csv('data/RP_n1_Asite_z_displacements.csv')
n2_z_disp = pd.read_csv('data/RP_n2_Asite_z_displacements.csv')
n3_z_disp = pd.read_csv('data/RP_n3_Asite_z_displacements.csv')
n4_z_disp = pd.read_csv('data/RP_n4_Asite_z_displacements.csv')
n5_z_disp = pd.read_csv('data/RP_n5_Asite_z_displacements.csv')
n6_z_disp = pd.read_csv('data/RP_n6_Asite_z_displacements.csv')
```

```python
colors = plt.cm.coolwarm(np.linspace(0,1,6))
fig, ax = plt.subplots(2,3, figsize=(6,4), dpi=200)
ax[0,0].plot(n1_z_disp.temp, n1_z_disp.z_diff_layer0, color= colors[0], label = 'Layer1')
ax[0,0].plot(n1_z_disp.temp, n1_z_disp.z_diff_layer1, color= colors[0], label = 'Layer2')

ax[0,1].plot(n2_z_disp.temp, n2_z_disp.z_diff_layer0, color= colors[1], label = 'Layer1')
ax[0,1].plot(n2_z_disp.temp, n2_z_disp.z_diff_layer1, color= colors[1], label = 'Layer2', alpha = 0.5)
ax[0,1].plot(n2_z_disp.temp, n2_z_disp.z_diff_layer2, color= colors[1], label = 'Layer3')

ax[0,2].plot(n3_z_disp.temp, n3_z_disp.z_diff_layer0, color= colors[2], label = 'Layer1')
ax[0,2].plot(n3_z_disp.temp, n3_z_disp.z_diff_layer1, color= colors[2], label = 'Layer2',alpha = 0.5)
ax[0,2].plot(n3_z_disp.temp, n3_z_disp.z_diff_layer2, color= colors[2], label = 'Layer3',alpha = 0.5)
ax[0,2].plot(n3_z_disp.temp, n3_z_disp.z_diff_layer3, color= colors[2], label = 'Layer4')

ax[1,0].plot(n4_z_disp.temp, n4_z_disp.z_diff_layer0, color= colors[3], label = 'Layer1')
ax[1,0].plot(n4_z_disp.temp, n4_z_disp.z_diff_layer1, color= colors[3], label = 'Layer2', alpha = 0.7)
ax[1,0].plot(n4_z_disp.temp, n4_z_disp.z_diff_layer2, color= colors[3], label = 'Layer3', alpha = 0.2)
ax[1,0].plot(n4_z_disp.temp, n4_z_disp.z_diff_layer3, color= colors[3], label = 'Layer4', alpha = 0.7)
ax[1,0].plot(n4_z_disp.temp, n4_z_disp.z_diff_layer4, color= colors[3], label = 'Layer5')

ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer0, color= colors[4], label = 'Layer1')
ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer1, color= colors[4], label = 'Layer2', alpha = 0.7)
ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer2, color= colors[4], label = 'Layer3', alpha = 0.2)
ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer3, color= colors[4], label = 'Layer4', alpha = 0.2)
ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer4, color= colors[4], label = 'Layer5', alpha = 0.7)
ax[1,1].plot(n5_z_disp.temp, n5_z_disp.z_diff_layer5, color= colors[4], label = 'Layer6')

ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer0, color= colors[5], label = 'Layer1')
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer1, color= colors[5], label = 'Layer2', alpha = 0.7)
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer2, color= colors[5], label = 'Layer3', alpha = 0.5)
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer3, color= colors[5], label = 'Layer4', alpha = 0.2)
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer4, color= colors[5], label = 'Layer5', alpha = 0.5)
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer5, color= colors[5], label = 'Layer6', alpha = 0.7)
ax[1,2].plot(n6_z_disp.temp, n6_z_disp.z_diff_layer6, color= colors[5], label = 'Layer7')
ax[0,0].text(-0.2,1, 'a)', transform = ax[0,0].transAxes)
ax[0,1].text(-0.2,1, 'b)', transform = ax[0,1].transAxes)
ax[0,2].text(-0.2,1, 'c)', transform = ax[0,2].transAxes)
ax[1,0].text(-0.2,1, 'd)', transform = ax[1,0].transAxes)
ax[1,1].text(-0.2,1, 'e)', transform = ax[1,1].transAxes)
ax[1,2].text(-0.2,1, 'f)', transform = ax[1,2].transAxes)
ax[0,0].set_title(r'$n=1$')
ax[0,1].set_title(r'$n=2$')
ax[0,2].set_title(r'$n=3$')
ax[1,0].set_title(r'$n=4$')
ax[1,1].set_title(r'$n=5$')
ax[1,2].set_title(r'$n=6$')
ax[0,0].set_xlim([0,1000])
ax[0,1].set_xlim([0,1000])
ax[0,2].set_xlim([0,1000])
ax[1,0].set_xlim([0,1000])
ax[1,1].set_xlim([0,1000])
ax[1,2].set_xlim([0,1000])
ax[0,0].set_ylabel(r'$\delta$ (\AA)')
ax[1,0].set_ylabel(r'$\delta$ (\AA)')
ax[1,0].set_xlabel('Temperature (K)') 
ax[1,1].set_xlabel('Temperature (K)') 
ax[1,2].set_xlabel('Temperature (K)') 
ax[0,0].legend(frameon=True)
ax[0,1].legend(frameon=True)
ax[0,2].legend(frameon=True)
ax[1,0].legend(frameon=True)
ax[1,1].legend(frameon=True)
ax[1,2].legend(frameon=True)
fig.tight_layout()
#plt.subplots_adjust(hspace=0)
plt.savefig('figs/SI_rumpling.png')
```
